```python
### 求出free_trial_after為yes的使用者 的前後頁面出現次數
import pandas as pd
import pandas_profiling as pp
from tqdm import tqdm

def find_frequency_of_moment_for_all_subscriber(batch_file_path):
    # 
    subscriber_info_batch_df = pd.read_csv(batch_file_path)
    subscriber_subscribe_yes = subscriber_info_batch_df[subscriber_info_batch_df["e_segment_map.subscriber"].str.count(r"yes")==1]
    
    subscriber_subscribe_yes_device_id_without_deplicates = subscriber_subscribe_yes["f_device_id"].drop_duplicates()
    subscriber_subscribe_yes_device_id_without_deplicates_list = list(subscriber_subscribe_yes_device_id_without_deplicates)
    
    col_e_key_operation = pd.DataFrame()
    for device_id in tqdm(subscriber_after_free_trial_yes_device_id_without_deplicates_list):
        col_e_key = subscriber_info_batch_df[subscriber_info_batch_df["f_device_id"].str.count(f"{device_id}")==1]["e_key"]
        col_operation = subscriber_info_batch_df[subscriber_info_batch_df["f_device_id"].str.count(f"{device_id}")==1]["e_segment_map.operation"]
        col_e_key_operation = pd.concat([col_e_key_operation ,col_e_key, col_operation], axis = 1)
    
    batch_number = batch_file_path.replace("batch", "")
    report = pp.ProfileReport(col_e_key_operation)
    report.to_file(f"output_col_e_key_operation_{batch_number}.html")
    
    

```

```python
import pandas as pd

def convert_sub_moment_data_into_value_count_and_dataframe(batch_file_path):
    subscriber_info_df = pd.read_csv(batch_file_path)
    
    ### 看 subscribe == yes (sub from moment) 有幾筆資料
    subscriber_subscriber_yes = subscriber_info_df[subscriber_info_df["e_segment_map.subscriber"].str.count(r"yes")==1]
    
    ### 提取出 subscribe == yes 的資料的 device_id
    subscriber_subcriber_yes_device_id_without_deplicates = subscriber_subscriber_yes["f_device_id"].drop_duplicates()
    subscriber_subscriber_yes_device_id_without_deplicates_list = list(subscriber_subcriber_yes_device_id_without_deplicates)
    
    df_value_counts_all = pd.DataFrame()
    for user in subscriber_subscriber_yes_device_id_without_deplicates_list:
        col_e_key = subscriber_info_df[subscriber_info_df["f_device_id"].str.count(f"{user}")==1]["e_key"]
        col_operation = subscriber_info_df[subscriber_info_df["f_device_id"].str.count(f"{user}")==1]["e_segment_map.operation"]
        col_e_key_operation = pd.concat([col_e_key, col_operation], axis = 1)
        
        summary_list = list()
        col_e_key_operation["final_value"] = 0
        for row in range(len(col_e_key_operation)):
            e_key_value = col_e_key_operation.iloc[row,]["e_key"]
            operation_value = col_e_key_operation.iloc[row,]["e_segment_map.operation"]
            final_value = f"{e_key_value} {operation_value}"
    
            col_e_key_operation.iloc[row, 2] = final_value
        
        value_counts = col_e_key_operation["final_value"].value_counts()
        df_value_counts = pd.DataFrame(value_counts)
        df_value_counts = df_value_counts.T
        
        df_value_counts_all = pd.concat([df_value_counts_all, df_value_counts], ignore_index=True)
        
    df_value_counts_all.to_csv("value_counts.csv")
```
