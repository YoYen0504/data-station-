```python
### 求出有按下訂閱按鈕前的導向頁面的出現次數 並回傳相對應的device_id
import pandas as pd

def subscribe_page_frequencies(batch_file_path):
    ### 讀取batch資料
    subscriber_operation_subscribe_df = pd.read_csv(batch_file_path)
    
    ### 進到某個頁面後 operation是subscribe的資料
    subscriber_operation_subscribe_yes_df = subscriber_operation_subscribe_df[subscriber_operation_subscribe_df["e_segment_map.operation"].str.count(r"subscribe")==1]
    
    ### 檢查訂閱前的頁面的出現次數
    subscriber_from_page_operation = subscriber_operation_subscribe_df[subscriber_operation_subscribe_df["e_segment_map.operation"].str.count(r"subscribe")==1]["e_key"] 
    print(subscriber_from_page_operation.value_counts())
    
    ### 提取出 after_free_trial == yes 的資料的 device_id
    subscriber_operation_subscribe_yes_device_id_without_deplicates = subscriber_operation_subscribe_yes_df["f_device_id"].drop_duplicates()
    subscriber_operation_subscribe_yes_device_id_without_deplicates = list(subscriber_operation_subscribe_yes_device_id_without_deplicates)
    
    return(subscriber_operation_subscribe_yes_device_id_without_deplicates)
```

```python
subscribe_page_frequencies("part_00096.csv")
```
