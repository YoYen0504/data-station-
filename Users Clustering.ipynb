```python
import pandas as pd
from sklearn import cluster, datasets, metrics
```

```python
value_counts_batch00000_df = pd.read_csv("value_counts_batch00000.csv")
value_counts_batch00032_df = pd.read_csv("value_counts_batch00032.csv")
value_counts_batch00064_df = pd.read_csv("value_counts_batch00064.csv")
value_counts_batch00096_df = pd.read_csv("value_counts_batch00096.csv")
value_counts_batch00160_df = pd.read_csv("value_counts_batch00160.csv")
value_counts_batch00192_df = pd.read_csv("value_counts_batch00192.csv")
value_counts_batch00224_df = pd.read_csv("value_counts_batch00224.csv")
value_counts_batch00256_df = pd.read_csv("value_counts_batch00256.csv")
value_counts_batch00288_df = pd.read_csv("value_counts_batch00288.csv")
```

```python
value_counts_all_df = pd.concat([value_counts_batch00000_df,
                                 value_counts_batch00032_df,
                                 value_counts_batch00064_df,
                                 value_counts_batch00096_df,
                                 value_counts_batch00160_df,
                                 value_counts_batch00192_df,
                                 value_counts_batch00224_df,
                                 value_counts_batch00256_df,
                                 value_counts_batch00288_df], ignore_index = True)
```

```python
value_counts_all_df.head()
```

```python
value_counts_all_df = value_counts_all_df.fillna(value = 0)
```

```python
value_counts_all_df.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
# KMeans 演算法
kmeans_fit = cluster.KMeans(n_clusters = 3).fit(value_counts_all_df)
```

```python
# 印出分群結果
cluster_labels = kmeans_fit.labels_
print("分群結果：")
print(cluster_labels)
print("---")
```

```python
len(cluster_labels)
```

```python
clustering_result = pd.DataFrame(cluster_labels)
```

```python
clustering_result
```

```python
value_counts_all_df.to_csv("value_counts_all.csv")
```

```python
value_counts_all_and_result = pd.concat([value_counts_all_df, clustering_result], axis = 1)
```

```python
value_counts_all_and_result
```

```python
value_counts_all_and_result.to_csv("value_counts_result.csv")
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

sil = []
kmax = 30

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
    k = KMeans(n_clusters = k).fit(value_counts_all_df)
    labels = k.labels_
    sil.append(silhouette_score(value_counts_all_df, labels, metric = 'euclidean'))
```
